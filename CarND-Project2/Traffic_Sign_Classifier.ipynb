````python
### Preprocess the data here.
### Feel free to use as many code cells as needed.
import cv2
from sklearn.preprocessing import LabelBinarizer
import tensorflow as tf

# tf.image.rgb_to_grayscale
def dataSetToGrayScale(dataset, sampleSize, imageShape):
    grayScale = np.empty([sampleSize, imageShape[0], imageShape[1]], dtype=dataset.dtype)
    for index in range(0, sampleSize):
        grayScale[index,:,:] = toGrayScale(dataset[index,:,:,:])
    grayScale = grayScale[:,:,:,np.newaxis]
    return grayScale

def toGrayScale(img):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# tf.one_hot
def oneHotEncoding(labels):
    encoder = LabelBinarizer()
    oneHotEncodingLabels= encoder.fit_transform(labels)

    return oneHotEncodingLabels

# Normalization
def normalizeData(img):
    outMin = 0.1
    outMax = 0.9
    greyscaleMin = 0
    greyscaleMax = 255
    
    slope = (outMax - outMin)/( greyscaleMax - greyscaleMin )
    b = outMin
    return slope*(img.astype(np.float32) - greyscaleMin) + outMin
````

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.
from sklearn.utils import shuffle
from tensorflow.contrib.layers import flatten

EPOCHS = 100
BATCH_SIZE = 128
dropout = 0.75   

def LeNet(x):
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1

    # Layer 1: Convolutional. Input = 32x32x1 with grayScale input channel  Output = 28x28x16.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 3, 16), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(16))
    conv1   = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID') + conv1_b

    # TODO: Activation.
    conv1 = tf.nn.relu(conv1)
    # Dropout
    conv1 = tf.nn.dropout(conv1,dropout)
    # TODO: Pooling. Input = 28x28x16. Output = 14x14x6.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # TODO: Layer 2: Convolutional. Output = 10x10x32.
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 16, 32), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(32))
    conv2   = tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='VALID') + conv2_b
    # TODO: Activation.
    conv2 = tf.nn.relu(conv2)
    # TODO: Pooling. Input = 10x10x16. Output = 5x5x16.
    # Dropout
    conv2 = tf.nn.dropout(conv2,dropout)
    # Pooling
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    # TODO: Flatten. Input = 5x5x16. Output = 800.
    fc0   = flatten(conv2)

    # TODO: Layer 3: Fully Connected. Input = 800. Output = 400.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(800, 400), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(400))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    # TODO: Activation.
    fc1    = tf.nn.relu(fc1)
    # Dropout
    fc1 = tf.nn.dropout(fc1,dropout)

    # TODO: Layer 4: Fully Connected. Input = 400. Output = 120.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(400, 120), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(120))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    # TODO: Activation.
    fc2    = tf.nn.relu(fc2)
    # Dropout
    fc2 = tf.nn.dropout(fc2,dropout)
    # TODO: Layer 5: Fully Connected. Input = 120. Output = 43.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(120, 43), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(43))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    return logits
````

**Answer:**
Since we know that for this lab, the input data are all images, a Convolutional Neural Network would be the best choice.
Also, I impleted a 2 times 2 max pooling layer for each conv layer. The purpose of using pooling layer is to reduce the spatial size of the representation, to reduce the number of parameters and amount of computation in the network, and hence to also control overfitting.
In order to prevent overfitting, I implemented a dropout layer. The dropout layer can reduce overfitting by preventing complex co-adaptations on training data.
In order to make an activation function, I chose to use relu(rather than sigmoid), because the relu gives a constant value for gradient which reduced the likelihood of the gradient to vanish.

Overall structure details:
- Convolutional Layer 1: 5x5 with 16 depth and 2x2 max pooling
- Convolutional Layer 2: 5x5 with 32 depth and 2x2 max pooling
- Fully Connected Layer: 2 layers
- Output layer: 43 neurons 
- Activation: relu
- relu followed by Dropout

````python
### Train your model here.
### Feel free to use as many code cells as needed.
````

````python
# Graph input
x = tf.placeholder(tf.float32, (None, 32, 32,3))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, 43)
keep_prob = tf.placeholder(tf.float32)
print(one_hot_y)
````

````python
rate = 0.001

logits = LeNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
import time


accuracyArray = []
with tf.Session() as sess:
    start_time = time.strftime("%c")
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)

# Data to gray scale
#     X_train = X_train_gray
#     X_train = X_train[:,:,:,np.newaxis]
    
#     X_valid = X_valid_gray
#     X_valid = X_valid[:,:,:,np.newaxis]
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        accuracyArray.append(validation_accuracy)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print() 
    
    # Test Data
    correct_prediction = evaluate(X_test, y_test)
    print("Test Accuracy= {:.3f}".format(correct_prediction))
    
    save_path = saver.save(sess, "Data/data.ckpt")
    print("Model saved in file: %s" % save_path)
    end_time = time.strftime("%c")
plt.plot(accuracyArray)
plt.ylabel('Accuracy')
plt.xlabel('EPOCH')
plt.show()
print("--------Training Started----------",start_time)
print("--------Training Ended----------",end_time)
````
